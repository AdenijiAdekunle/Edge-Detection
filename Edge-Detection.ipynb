```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def simple_edge_detection(image):
    edges_detected = cv2.Canny(image, 150, 150)
    images = [image, edges_detected]
    location = [121, 122]
    for loc, edge_image in zip(location, images):
        plt.subplot(loc)
        plt.imshow(edge_image, cmap='gray')
        cv2.imwrite('car.png', edges_detected)
        plt.show()
```

```python
img = cv2.imread('car.jpg', 0)
```

```python
simple_edge_detection(img)
```

```python
def edge_detection(image):
    edges_detected = cv2.Canny(image, 200, 150)
    images = [image, edges_detected]
    location = [121, 122]
    for loc, edge_image in zip(location, images):
        plt.subplot(loc)
        plt.imshow(edge_image, cmap='gray')
        cv2.imwrite('stairs.png', edges_detected)
        plt.show()
```

```python
pic = cv2.imread('stairs.jpg', 0)
```

```python
edge_detection(pic)
```
